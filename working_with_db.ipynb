### Working with sqlite database

```python
import sqlite3
```

```python
try:
    with sqlite3.connect("test.db") as conn:
        print("Sql connection successful!")
        
except sqlite3.OperationalError as e:
    print("Failed to open database:", e)
```

```python
conn.close()
```

```python
create_table_sql = """CREATE TABLE IF NOT EXISTS user_details(
    id INTEGER PRIMARY KEY,
    first_name text NOT NULL,
    last_name text NOT NULL,
    email text NOT NULL
)
"""
ins_table_sql = """INSERT INTO user_details(first_name,last_name,email)
values(?,?,?)"""

ins_table_values = ("John","Smith","john.smith@abc.com")
```

```python
try:
    with sqlite3.connect("test.db") as conn:
        cur = conn.cursor()
        cur.execute(create_table_sql)
        print("User table created...")
except sqlite3.OperationalError as e:
    print("Sql error:", e)
```

```python
try:
    cur.execute(ins_table_sql,ins_table_values)
    print("User record entered....")
except sqlite3.OperationalError as e:
    print("Exception with sql insert:",e)
```

```python
ins_table_values = ("Mike","Ray","mike.ray@abc.com")
try:
    cur.execute(ins_table_sql,ins_table_values)
    print("User record entered....")
except sqlite3.OperationalError as e:
    print("Exception with sql insert:",e)
```

```python
ins_table_values = ("Jack","Robert","jack.robert@abc.com")
try:
    cur.execute(ins_table_sql,ins_table_values)
    print("User record entered....")
except sqlite3.OperationalError as e:
    print("Exception with sql insert:",e)
```

```python
scur = conn.cursor()
scur.execute("SELECT * FROM user_details")
rows = scur.fetchall()
conn.close()

for row in rows:
    print(row)
```

```python
create_table = """CREATE TABLE IF NOT EXISTS HISTORY_TBL(
    id INTEGER PRIMARY KEY,
    user_question text NOT NULL,
    pros_list text,
    cons_lis text,
    ai_analysis text
)
"""

try:
    with sqlite3.connect("decision_history.db") as conn:
        connection = conn.cursor()
        connection.execute(create_table)
        print("History table created....")
except sqlite3.OperationalError as e:
    print("Exception with table creation:", e)
        
```

```python
conn.close()
```

```python
try:
    with sqlite3.connect("decision_history.db") as conn:
        con = conn.cursor()
        con.execute("SELECT * FROM HISTORY_TBL")
        rows = con.fetchall()
except sqlite3.OperationalError as er:
    print("Error getting the data from table", er)

for row in rows:
    print(row)
```
